```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img=cv2.imread('/content/cat.jpeg')
```

```python
cv2_imshow(img)
cv2.waitKey(0)
```

```python
img_resize=cv2.resize(img,(800,800))
cv2_imshow(img_resize)
cv2.waitKey(0)
```

```python
img_crop = img[100:300, 200:500]  # Crop from row 100 to 300 and column 200 to 500
cv2_imshow(img_crop)  # Display the cropped image directly
cv2.waitKey(0)
```

```python
threshold_img=cv2.threshold(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY),125,255,cv2.THRESH_BINARY)
cv2_imshow(threshold_img[1])
cv2.waitKey(0)
```
