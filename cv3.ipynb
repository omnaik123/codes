```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/cat.jpeg')
```

```python
cv2_imshow(img)
cv2.waitKey(0)
```

```python
def draw_line(img,pt1,pt2,color,thickness,style=cv2.LINE_AA):
  cv2.line(img,pt1,pt2,color,thickness,style)
```

```python
def draw_text(imag,text,pt,color,thickness):
  cv2.putText(imag,text,pt,cv2.FONT_HERSHEY_SIMPLEX,1,color,thickness)
```

```python
def draw_circle(imag,center,radius,color,thickness):
  cv2.circle(imag,center,radius,color,thickness)
```

```python
def draw_rectangle(imag,pt1,pt2,color,thickness):
  cv2.rectangle(imag,pt1,pt2,color,thickness)
```

```python
def draw_ellipse(imag,center,axes,angle,start,end,color,thickness):
  cv2.ellipse(imag,center,axes,angle,start,end,color,thickness)
```

```python
draw_line(img,(100,100),(200,200),(255,0,0),2)
draw_text(img, 'Hello', (50,50),(0,255,0),2)
draw_circle(img,(100,100),50,(0,0,255),2)
draw_rectangle(img,(100,100),(200,200),(0,255,0),2)
draw_ellipse(img,(100,100),(50,20),0,0,360,(255,0,0),2)
```

```python
cv2_imshow(img)
cv2.waitKey(0)
```
